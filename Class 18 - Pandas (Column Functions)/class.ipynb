### Apply Custom Function on Columns (Series Data)

```python
def grade(per:int)->str:
    if per >= 80:
        return "A+"
    elif per >= 70:
        return "A"
    elif per >= 60:
        return "B"
    elif per >= 50:
        return "C"
    elif per >= 40:
        return "D"
    elif per >= 33:
        return "E"
    else:
        return "F"
```

```python
df['Grade']=df['Percentage'].apply(grade)
df
```

```python
def fn_obtain(s1 : int , s2 : int, s3: int, s4: int , s5: int) -> tuple[int,int,float,str]:
    total : int = 500
    obtained : int = s1 + s2 + s3 + s4 + s5
    per : float = obtained / total *100
    grade : str = ""
    if per >= 80:
        grade = "A+"
    elif per >= 70:
        grade = "A"
    elif per >= 60:
        grade = "B"
    elif per >= 50:
        grade = "C"
    elif per >= 40:
        grade = "D"
    elif per >= 33:
        grade = "E"
    else:
        grade = "F"
    return total ,obtained ,per ,grade

df1[["Total","Obtained Marks","Percentage","Grade"]] = df1[['s1','s2','s3','s4','s5']].apply(lambda x :fn_obtain(*x),axis=1,result_type='expand')
df1
```

# Map Function For Remarks

```python
from typing import Dict


remarks : Dict[str,str] = {
    "A+":"Excellent",
    "A":"Very Good",
    "B": " Good",
    "C": "Fair",
    "D": "Poor",
    "E": "Very Poor",
    "F": "Fail"
}
```

```python
df["Remarks"]= df.Grade.map(remarks)
df
```
