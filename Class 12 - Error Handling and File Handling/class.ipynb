Use PyPDF2 to show PDF files


````python
from typing import List
import PyPDF2

def readpdf(file_path:str)->List[str]:
    with open(file_path,'rb') as file:
        reader : PyPDF2 = PyPDF2.PdfReader(file)
        text_content: List[str] =[page.extract_text() for page in reader.pages]
        return text_content
    
page = readpdf("./Pakistan.pdf")
page
````

Using matplotlib.pyplot to show audio

````python
import matplotlib.pyplot as plt
import numpy as np
import wave

# Replace "./your_audio.wav" with the path to your audio file.
audio_file = wave.open("./audio.wav", 'rb')

signals = audio_file.readframes(-1)
signals = np.frombuffer(signals, dtype=np.int16)

plt.figure(1)
plt.title("Signal waves....")
plt.plot(signals)
plt.show()

audio_file.close()  # Close the audio file when you're done.
````
